### find

```python
import requests
from bs4 import BeautifulSoup

url = "https://webscraper.io/test-sites/e-commerce/allinone"

r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")

price = (soup.find("h4",{"class": "float-end price card-title pull-right"}))
dec = soup.find("p",{'class': "description card-text"})
a = soup.find("p",class_ = "description")
print(a.string)
print(dec.string)
# find fn is limited to first instance 
```

### find_all

```python
import requests
from bs4 import BeautifulSoup

url = "https://webscraper.io/test-sites/e-commerce/allinone"

r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")
price = soup.header.div.h4
prices = soup.find_all(price , class_ = "float-end price card-title pull-right")
for i in prices:
    print(i.text)
```

### find_all regex

```python
import requests
from bs4 import BeautifulSoup
import re #regular expression

url = "https://webscraper.io/test-sites/e-commerce/allinone"

r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")

data = soup.find_all(['p','h4','a'])
print(data)
```

```python
import requests
from bs4 import BeautifulSoup
import re #regular expression

url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"

r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")

# data = soup.find_all(string = ('Galaxy'))
data = soup.find_all(string = re.compile('Idea'))
print(data)
```
