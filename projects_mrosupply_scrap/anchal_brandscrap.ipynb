```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
import random

urls = pd.read_csv("mroscrapwork (1).csv")
url_list = []
status_code_list = []
title_list = []
model_list = []
spec_key_list = []
spec_value_list = []
breadcrumb_list = []
data_link_list = []

for i, url in enumerate(urls['URLs'].iloc[470:]):
    r = requests.get(url)
    status_code = r.status_code  # Store the status code
    if status_code == 200:
        soup = BeautifulSoup(r.text, "lxml")
        print(f'scraping {i} {url}')
        
        # Find title
        title_element = soup.find("h1", class_="product-card__title")
        title = title_element.text if title_element else None
        title_list.append(title)

        # Find model
        model_element = soup.find("p", class_="modelNo")
        model = model_element.text if model_element else None
        model_list.append(model)

        # Find breadcrumb
        breadcrumbs = [item.get_text(strip=True) for item in soup.select('.m-breadcrumbs-list--item [itemprop="name"]')]
        breadcrumb_list.append(breadcrumbs)

        # find datasheet
        datasheet = soup.find("div", class_="documents--item")
        sheet = datasheet.find("a")['href'] if datasheet else None
        data_link_list.append(sheet)

        # Find specifications
        lis = soup.find("div", class_='m-accordion js-prod-accordion')
        if lis:
            spec_key = [i.text for i in lis.find_all("p", class_="key")]
            spec_value = [i.text for i in lis.find_all("p", class_="value")]
        else:
            spec_key = []
            spec_value = []

        spec_key_list.append(spec_key)
        spec_value_list.append(spec_value)
        url_list.append(url)
        status_code_list.append(status_code)

        # Introduced a random delay between 2 and 5 seconds
        time.sleep(random.uniform(2, 5))

# Create DataFrame after the loop
df3 = pd.DataFrame({
    "urls": url_list,
    "status_code": status_code_list,
    "tinken_title": title_list,
    "tinken_model": model_list,
    "key": spec_key_list,
    "value": spec_value_list,
    "breadcrumb": breadcrumb_list,
    "data_sheet" : data_link_list
})

df3.to_csv("anchal_scrap2.csv")
```
